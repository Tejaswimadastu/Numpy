Data Engineer

```python
import numpy as np

# -------------------------------
# 1. Create farm arrays
# -------------------------------
farmA = np.array([[10,12,11],[9,11,10],[12,13,14],[8,7,9]])
farmB = np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farmC = np.array([[15,16,14],[14,13,15],[16,17,18],[12,11,13]])

# Convert to float to allow decimal operations
production_all = np.array([farmA, farmB, farmC], dtype=float)

# -------------------------------
# 2. Array attributes
# -------------------------------
print("Production array:\n", production_all)
print("Shape ", production_all.shape)
print("Size ", production_all.size)
print("Data type ", production_all.dtype)
print("Dimension ", production_all.ndim)

# -------------------------------
# 3. Per-farm totals & stats
# -------------------------------
total_farm = production_all.sum(axis=(1,2))
mean_farm = production_all.mean(axis=(1,2))
std_farm = production_all.std(axis=(1,2))
var_farm = production_all.var(axis=(1,2))
print("Total per farm:", total_farm)
print("Mean per farm:", mean_farm)
print("Std per farm:", std_farm)
print("Var per farm:", var_farm)

# -------------------------------
# 4. Weighted per-panel energy
# -------------------------------
hour_weights = np.array([0.5, 1.0, 0.75])
weighted_per_panel = np.dot(production_all, hour_weights)
print("Weighted per panel:\n", weighted_per_panel)

# -------------------------------
# 5. Calibration offset for Farm B
# -------------------------------
production_all[1] += 0.5
print("After calibration offset Farm B:\n", production_all[1])

# -------------------------------
# 6. Z-score normalization per farm
# -------------------------------
mean_farm_keep = production_all.mean(axis=(1,2), keepdims=True)
std_farm_keep = production_all.std(axis=(1,2), keepdims=True)
normalized = (production_all - mean_farm_keep) / std_farm_keep
print("Normalized Farm A:\n", normalized[0])

# -------------------------------
# 7. Element-wise ufuncs
# -------------------------------
# Exponential family
exp_vals = np.exp(normalized)
expm1_vals = np.expm1(normalized)
log_vals = np.log(exp_vals)
log10_vals = np.log10(exp_vals)
log2_vals = np.log2(exp_vals)

# Trigonometric
sin_vals = np.sin(normalized)
cos_vals = np.cos(normalized)
tan_vals = np.tan(normalized)
arcsin_vals = np.arcsin(np.clip(normalized/np.max(np.abs(normalized)), -1,1))  # avoid invalid
arccos_vals = np.arccos(np.clip(normalized/np.max(np.abs(normalized)), -1,1))
arctan_vals = np.arctan(normalized)
degrees_vals = np.degrees(arctan_vals)
radians_vals = np.radians(degrees_vals)

# Rounding
rounded_vals = np.round(normalized)
floored_vals = np.floor(normalized)
ceiled_vals = np.ceil(normalized)
truncated_vals = np.trunc(normalized)

# Comparison & logical
high_output = production_all > 20
even_hours_mask = (np.arange(3) % 2 == 0).reshape(1,1,3)
logical_and_mask = np.logical_and(high_output, even_hours_mask)
logical_or_mask = np.logical_or(high_output, even_hours_mask)
logical_not_mask = np.logical_not(high_output)
logical_xor_mask = np.logical_xor(high_output, even_hours_mask)

# Bitwise ufuncs
alerts = np.array([
    [[1,0,1],[0,0,1],[1,1,0],[0,1,0]],  # Farm A
    [[1,0,1],[0,0,1],[1,1,0],[0,1,0]],  # Farm B
    [[1,0,1],[0,0,1],[1,1,0],[0,1,0]]   # Farm C
], dtype=int)
bitwise_and_vals = np.bitwise_and(alerts, 1)
bitwise_or_vals = np.bitwise_or(alerts, 1)
bitwise_invert_vals = np.bitwise_not(alerts)

# Arithmetic ufuncs
added_vals = normalized + 1
subtracted_vals = normalized - 1
multiplied_vals = normalized * 2
divided_vals = normalized / 2
powered_vals = np.power(normalized, 2)
mod_vals = np.mod(normalized, 2)

# Aggregations over ufunc outputs
sum_exp = np.sum(exp_vals)
mean_sin = np.mean(sin_vals)
print("Sum of exp(normalized):", sum_exp)
print("Mean of sin(normalized):", mean_sin)

# -------------------------------
# 8. Panel totals & maintenance check
# -------------------------------
panel_totals_per_farm = production_all.sum(axis=2)
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=1)
print("Panel totals per farm:\n", panel_totals_per_farm)
print("Needs maintenance:", needs_maintenance)

# -------------------------------
# 9. Final KPIs
# -------------------------------
overall_total = production_all.sum()
peak_hour_value = production_all.max()
peak_idx = np.unravel_index(np.argmax(production_all), production_all.shape)
average_per_panel = production_all.mean(axis=2)
print("Overall total energy:", overall_total)
print("Peak hour value:", peak_hour_value)
print("Peak produced by farm/panel:", peak_idx)
print("Average per panel:\n", average_per_panel)
```
